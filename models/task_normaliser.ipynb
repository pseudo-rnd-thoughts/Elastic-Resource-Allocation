```python
import pandas as pd
```

```python
df = pd.read_csv('reduced_task_events.csv')
```

```python
df
```

```python
nan_df = df[(df['resource request for CPU cores'] > 0) &
            (df['resource request for RAM'] > 0) &
            (df['resource request for local disk space'] > 0)]
```

```python
grouped_df = nan_df.groupby(['job ID', 'priority', 'resource request for CPU cores', 'resource request for RAM', 
                             'resource request for local disk space'], as_index=False).size().reset_index(name='Count')
```

```python
grouped_df
```

```python
grouped_df.sort_values(by='Count', ascending=False)
```

```python
top_df = grouped_df[grouped_df['Count'] >= 1000]
```

```python
top_df
```

```python
total = top_df['Count'].sum()
total
```

```python
prob_df = top_df
prob_df['Probablity'] = prob_df['Count'].apply(lambda count: count / total)
prob_df
```

```python
min_cpu_core = prob_df['resource request for CPU cores'].min()
min_cpu_core
```

```python
min_ram = prob_df['resource request for RAM'].min()
min_ram
```

```python
min_disk_space = prob_df['resource request for local disk space'].min()
min_disk_space
```

```python
normalised_df = prob_df
normalised_df['normalised CPU cores'] = prob_df['resource request for CPU cores'].apply(lambda cpu_core: cpu_core / min_cpu_core)
normalised_df['normalised RAM'] = prob_df['resource request for RAM'].apply(lambda ram: ram / min_ram)
normalised_df['normalised Disk Space'] = prob_df['resource request for local disk space'].apply(lambda disk_space: disk_space / min_disk_space)
normalised_df
```

```python
normalised_df.to_csv("normalised_task_events.csv", index=False)
```
