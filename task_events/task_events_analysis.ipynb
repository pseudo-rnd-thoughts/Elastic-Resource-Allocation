```python
import csv
from tqdm import tqdm
import os
import pandas as pd
import random
```

```python
filenames = [f for f in os.listdir('.') if '.csv' in f]
random.shuffle(filenames)
data = []
```

```python
filename = filenames[0]
print(filename)
with open(filename) as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row)
```

```python
df = pd.DataFrame(data, columns=['timestamp', 'missing info', 'job ID', 'task index', 'machine ID',
                                 'event type', 'user name', 'scheduling class', 'priority', 
                                 'CPU cores', 'RAM', 'disk space', 'different-machine constraint'])
```
